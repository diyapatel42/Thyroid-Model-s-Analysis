```r
# Column names are from file allhyper.names
hypo <- read.table("allhypo.data", sep = ",")
hypo <- hypo %>%
    rename("Age"="V1", "Sex"="V2", "On_thyroxine"="V3", "Query_on_thyroxine"="V4", "On_anti_medi"="V5", "Sick"="V6", "Pregnant"="V7", "Surgery"="V8", "I131_treatment"="V9", "Query_hypo"="V10",
           "Query_hyper"="V11", "Lithium"="V12", "Goitre"="V13", "Tumor"="V14", "Hypopituitary"="V15", "Psych"="V16", "TSH_measured"="V17", "TSH"="V18", "T3_measured"="V19", "T3"="V20",
           "TT4_measured"="V21", "TT4"="V22", "T4U_measured"="V23", "T4U"="V24", "FTI_measured"="V25", "FTI"="V26", "TBG_measured"="V27", "TBG"="V28", "Source"="V29", "Hypo"="V30")
head(hypo)
```

```r
# Files with .data are training sets (2800 obs), and files with .test are testing sets (972 obs)
train <- hypo
str(train)
```

```r
test <- read.table("allhypo.test", sep = ",")
test <- test %>%
    rename("Age"="V1", "Sex"="V2", "On_thyroxine"="V3", "Query_on_thyroxine"="V4", "On_anti_medi"="V5", "Sick"="V6", "Pregnant"="V7", "Surgery"="V8", "I131_treatment"="V9", "Query_hypo"="V10",
           "Query_hyper"="V11", "Lithium"="V12", "Goitre"="V13", "Tumor"="V14", "Hypopituitary"="V15", "Psych"="V16", "TSH_measured"="V17", "TSH"="V18", "T3_measured"="V19", "T3"="V20",
           "TT4_measured"="V21", "TT4"="V22", "T4U_measured"="V23", "T4U"="V24", "FTI_measured"="V25", "FTI"="V26", "TBG_measured"="V27", "TBG"="V28", "Source"="V29", "Hypo"="V30")
head(test)
str(test)
```

```r
# By taking others' research as a reference, variable Source is irrelevant
train <- train %>%
    dplyr::select(-Source)
test <- test %>%
    dplyr::select(-Source)
str(train)
str(test)
```

```r
# Hyper shows the classification, ".|" and numbers after this indexing the obs, meaningless, can be removed
for (i in 1:2800){
    text <- train$Hypo[i]
    rex <- "\\..*"
    train$Hypo[i] <- str_replace(text, rex, "")
}
```

```r
for (i in 1:972){
    text <- test$Hypo[i]
    rex <- "\\..*"
    test$Hypo[i] <- str_replace(text, rex, "")
}
```

```r
# See if any 
for (i in 2:29){
    print(unique(train[[i]]))
}
```

```r
for (i in 2:29){
    print(unique(test[[i]]))
}
```

```r
# We don't have any obs in variables TBG_measured and TBG, dropped
train <- train %>%
    select(-c(TBG_measured, TBG))
str(train)
```

```r
test <- test %>%
    select(-c(TBG_measured, TBG))
str(test)
```

```r
# Changing variables to dummies and assign NA to null values
for (i in 1:2800){
    for (j in 3:26){
        if (train[i,j] == "f"){
            train[i,j] = 0
        } else if (train[i,j] == "t"){
            train[i,j] = 1
        } else if (train[i,j] == "?"){
            train[i,j] = NA
        }
    }
}
```

```r
for (i in 1:972){
    for (j in 3:26){
        if (test[i,j] == "f"){
            test[i,j] = 0
        } else if (test[i,j] == "t"){
            test[i,j] = 1
        } else if (test[i,j] == "?"){
            test[i,j] = NA
        }
    }
}
```

```r
# Changing Sex to a dummy variable where 1 means female and 0 means male
for (i in 1:2800){
    if (train$Sex[i] == "F"){
        train$Sex[i] = 1
    } else if (train$Sex[i] == "M"){
        train$Sex[i] = 0
    }
}
train <- train %>%
    rename("Female" = "Sex")
```

```r
for (i in 1:972){
    if (test$Sex[i] == "F"){
        test$Sex[i] = 1
    } else if (test$Sex[i] == "M"){
        test$Sex[i] = 0
    }
}
test <- test %>%
    rename("Female" = "Sex")
```

```r
# Our class is one when the obs has hypothyroid or diagnoses diseases related to/ might lead to hypothyroid
for (i in 1:2800){
    if (train$Hypo[i] == "negative"){
        train$Hypo[i] = 0
    } else if (train$Hypo[i] != "negative"){
        train$Hypo[i] = 1
    }
}
train$Hypo <- as.numeric(train$Hypo)
unique(train$Hypo)
str(train)
```

```r
for (i in 1:972){
    if (test$Hypo[i] == "negative"){
        test$Hypo[i] = 0
    } else if (test$Hypo[i] != "negative"){
        test$Hypo[i] = 1
    }
}
test$Hypo <- as.numeric(test$Hypo)
unique(test$Hypo)
str(test)
```

```r
# How many obs have null values in these variables with a classification of 1
# If the amount is not very large comparing to total number of class 1 in our data, probably we could drop
# them
measures <- c(18, 20, 22, 24, 26)
columns <- c("TSH", "T3", "TT4", "T4U", "FTI")
number <- rep(0, 26)
for (i in measures){
    n <- train %>%
        filter(is.na(train[[i]]) == TRUE & train$Hypo == 1) %>%
        nrow()
    number[i] <- n
}
counts <- rep(0, 5)
for (i in 1:5){
    counts[i] <- number[measures[i]]
}
(train.null <- data.frame(Columns = columns, Counts = counts))

train %>%
    filter(is.na(train[[24]]) == TRUE & is.na(train[[26]]) == TRUE & train$Hypo == 1) %>%
    nrow()
train %>%
    filter(is.na(train[[24]]) == TRUE & is.na(train[[20]]) == TRUE & train$Hypo == 1) %>%
    nrow()
train %>%
    filter(is.na(train[[24]]) == TRUE & is.na(train[[22]]) == TRUE & train$Hypo == 1) %>%
    nrow()
train %>%
    filter(is.na(train[[20]]) == TRUE & is.na(train[[22]]) == TRUE & train$Hypo == 1) %>%
    nrow()

# If choose to drop NA obs in these variables from the train set, the number of class 1 would lose 57 obs
```

```r
measures <- c(18, 20, 22, 24, 26)
columns <- c("TSH", "T3", "TT4", "T4U", "FTI")
number <- rep(0, 26)
for (i in measures){
    n <- test %>%
        filter(is.na(test[[i]]) == TRUE & test$Hypo == 1) %>%
        nrow()
    number[i] <- n
}
counts <- rep(0, 5)
for (i in 1:5){
    counts[i] <- number[measures[i]]
}
(test.null <- data.frame(Columns = columns, Counts = counts))

test %>%
    filter(is.na(test[[20]]) == TRUE & is.na(test[[26]]) == TRUE & test$Hypo == 1) %>%
    nrow()

# If choose to drop NA obs in these variables from the test set, the number of class 1 would lose 11 obs
```
